```python
from json import load
import os
```

```python
jd_loc = r'json'
cd_loc = r'csv'
cars = []
fishs = []
houses = []
trees = []
bicycles = []
guitars = []
pencils = []
clocks = []

items = {
    'car': cars,
    'fish': fishs,
    'house': houses,
    'tree': trees,
    'bicycle': bicycles,
    'guitar': guitars,
    'pencil': pencils,
    'clock': clocks,
}
```

```python
def get_n_paths(shape):
    return len(shape)

def get_n_points(shape):
    return sum( len(path) for path in shape )

info = {
    'n_paths': get_n_paths,
    'n_points' : get_n_points,
}
```

```python
def turn_jd_to_cd(jf_loc):
    with open(jf_loc, 'r') as file:
        jf = load(file)
        for item, l in items.items():
            data = [] 
            shape = jf.get('drawings').get(item)
            for _, info_func in info.items():
                data.append(info_func(shape))
                
            #data.append(item) # class
            l.append(data)
```

```python
jf_names = os.listdir(jd_loc)[:]
exceptions = []
for n, jf_name in enumerate(jf_names):
    jf_loc = os.path.join(jd_loc, jf_name)
    try : 
        print(f'{n+1}/{len(jf_names)}', end='\r')
        turn_jd_to_cd(jf_loc)
    except : 
        print("Exception :", f'{n+1}/{len(jf_names)} : {jf_name}')
        exceptions.append(jf_name)
```
